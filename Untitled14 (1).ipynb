```python
#1: Loan Approval Prediction

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import accuracy_score, r2_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import mean_absolute_error


X = [
    [25000, 550],
    [35000, 650],
    [45000, 700],
    [20000, 500],
    [50000, 750]
]

# Labels
y = [
    "Reject",
    "Approve",
    "Approve",
    "Reject",
    "Approve"
]

# Create KNN model with K = 3
knn = KNeighborsClassifier(n_neighbors=3)

# Train the model
knn.fit(X, y)

# New applicant
new_applicant = [[40000, 680]]

# Predict loan status
prediction = knn.predict(new_applicant)

print("Predicted Loan Status:", prediction[0])

# Show the 3 nearest neighbors
distances, indices = knn.kneighbors(new_applicant)

print("\nNearest Neighbors:")
for i in range(len(indices[0])):
    idx = indices[0][i]
    print(
        f"Neighbor {i+1}: Income={X[idx][0]}, "
        f"Credit Score={X[idx][1]}, Loan={y[idx]}, "
        f"Distance={distances[0][i]:.2f}"
    )
```
