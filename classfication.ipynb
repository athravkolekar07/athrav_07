```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import accuracy_score, r2_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import mean_absolute_error


X = [
    [110, 90],
    [120, 110],
    [140, 150],
    [150, 170],
    [160, 180]
]

y = [
    "No",
    "No",
    "Yes",
    "Yes",
    "Yes"
]

# Create KNN model with K = 3
knn = KNeighborsClassifier(n_neighbors=3)

# Train the model
knn.fit(X, y)

# New patient data
new_patient = [[130, 130]]

# Predict disease status
prediction = knn.predict(new_patient)

print("Predicted Disease Status:", prediction[0])

# Display the 1 nearest neighbors
distances, indices = knn.kneighbors(new_patient)

print("\nNearest Neighbors:")
for i in range(len(indices[0])):
    idx = indices[0][i]
    print(
        f"Neighbor {i+1}: BP={X[idx][0]}, "
        f"Sugar={X[idx][1]}, Disease={y[idx]}, "
        f"Distance={distances[0][i]:.2f}"
    )
```
